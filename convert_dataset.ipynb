```python
import pandas as pd
import os
```

```python
def process_csv(path = os.getcwd()):
    df_list=[]
    file_list = os.listdir(path)
    file_list = [file for file in file_list if ('.csv' in file)]
    
    for file in file_list:
        print(file)
        
        df = pd.read_csv(file, encoding='cp1252')
        df = df.rename(columns={'Intermediate ': 'Intermediate'})
        df['Topic'] = file[:-4]
        df['Elementary'] = df['Elementary'].str.replace('\n', ' ')
        df['Intermediate'] = df['Intermediate'].str.replace('\n', ' ')
        df['Advanced'] = df['Advanced'].str.replace('\n', ' ')
        df_list.append(df[['Topic', 'Elementary', 'Intermediate', 'Advanced']])
    
    data = pd.concat(df_list)
    return data
```

```python
def process_txt():
    path = '/Users/Jialin/Downloads/OneStopEnglishCorpus-master/Texts-SeparatedByReadingLevel/'
    file_cat = ['Ele-Txt/', 'Adv-Txt/','Int-Txt/']
    df_list=[]
    
    for cat in file_cat:
        file_list = os.listdir(path + cat)
        file_list = [file for file in file_list if ('.txt' in file)]

        for file in file_list:
            #print(file)
            cate = file[-7:-4]
            df = pd.read_table(path + cat + file, sep='\n',header=None,names=['text'])
            df['cat'] = cate
            if ('WNL' in file[:-8]):
                df['topic'] = file[4:-8]
            else:
                df['topic'] = file[:-8]
            df_list.append(df)
    
    data = pd.concat(df_list)
    return data
```

```python
data = process_txt()
```
